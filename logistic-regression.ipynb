```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, classification_report
from scipy.io import mmread
```

```python
test_csv = pd.read_csv('data/Test.csv')
train_csv = pd.read_csv('data/Train.csv')

train_y = train_csv['label']
test_y = test_csv['label']

x_train_tf = mmread('data/x_train_tf.mtx')
x_test_tf = mmread('data/x_test_tf.mtx')
```

```python
logistic_regression = LogisticRegression()
logistic_regression.fit(x_train_tf, train_y)

y_pred = logistic_regression.predict(x_test_tf)
```

```python
matrix = confusion_matrix(test_y, y_pred)
plt.figure()
sns.heatmap(matrix, annot=True, fmt=".3f", linewidths=.5, square=True, cmap="Greens_r")
plt.ylabel('Actual class')
plt.xlabel('Predicted class')
plt.title('Logistic Regression', size=12)
plt.show()
```

```python
scores = classification_report(test_y, y_pred)
print(scores)
```
